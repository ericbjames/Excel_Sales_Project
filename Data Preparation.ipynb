```python
import pandas as pd
import os
import glob
```

```python
#Combining 12 individual month files

os.chdir("/Users/ericjames/Sales_Data")
extension="csv"
all_filenames=[file for file in glob.glob('*.{}'.format(extension))]
combined_csv=pd.concat([pd.read_csv(f) for f in all_filenames])
combined_csv.to_csv("all_data.csv",index=False,encoding="utf-8-sig")
```

```python
#Clean cells without data

all_data = pd.read_csv("all_data.csv")

drop_nan = all_data[all_data.isna().any(axis=1)]

all_data = all_data.dropna(how='all')
```

```python
#Create a separate month column

all_data = all_data[all_data['Order Date'].str[0:2] != "Or"]

all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
# Create an Order Sales column

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data['Order_Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

```python
# Separate the 'Order Date' Column into 'Date' & 'Hour'

all_data['Hour'] = all_data['Order Date'].str[-5:-3]
all_data['Date']= all_data['Order Date'].str[:9]
all_data.tail()
```

```python
# Separate 'Purchase Address' into separate columns

all_data[['Street','City','State_Zip']]=all_data['Purchase Address'].str.split(',',expand=True)

all_data['State_ab'] = all_data['State_Zip'].str[:3]
all_data['City_State'] = all_data['City']+ "," +all_data['State_ab']

all_data['Zip']=all_data['State_Zip'].str[4:]
all_data.tail()
```

```python
#Drop columns not used in final analysis by creating a new df and exportig the file

final_data = all_data.drop(['Street','State_Zip','Order Date','City','Zip'],axis=1,inplace=False)
final_data.tail()
```

```python
final_data.to_csv('/Users/ericjames/Sales_Data/final_data_export.csv', index=False)
```
